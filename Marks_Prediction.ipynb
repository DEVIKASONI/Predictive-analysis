```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
dataset=pd.read_csv('student.csv')
dataset.head()
```

```python
dataset.isnull().sum()
# no null values found
```

```python
dataset.shape
dataset.columns
```

```python
# changing categorical values to numeric values 
# creating dummy variables
# school, sex, address, family size, Pstatus, fedu, medu, schoolsup, famsup,paid, activities, nursery, higher,internet,romantic
```

```python
dummy=pd.get_dummies(dataset, columns=['school','sex','address','romantic','internet','higher','activities','nursery','famsize','Pstatus','schoolsup','famsup','paid'])
```

```python
dummy.head()
```

```python
pd.set_option('display.max_columns', None)
dummy.head()
```

```python
dummy=dummy.drop(columns=['school_MS','sex_M','address_U','romantic_no','internet_no','higher_no','activities_no','nursery_no','famsize_GT3','Pstatus_A','schoolsup_no','famsup_no','paid_no'],axis=1)
dummy.head()
```

```python
dummy=dummy.rename(columns={'school_GP':'school'})
dummy=dummy.rename(columns={'sex_F':'sex'})
dummy=dummy.rename(columns={'sex_F':'sex'})
dummy=dummy.rename(columns={'address_R':'address'})
dummy=dummy.rename(columns={'romantic_yes':'romantic'})
dummy=dummy.rename(columns={'internet_yes':'internet_yes'})
dummy=dummy.rename(columns={'higher_yes':'higher'})
dummy=dummy.rename(columns={'nursery_yes':'nursery'})
dummy=dummy.rename(columns={'Pstatus_T':'Pstatus'})
dummy=dummy.rename(columns={'famsize_LE3':'famsize'})
dummy=dummy.rename(columns={'schoolsup_yes':'schoolsup'})
dummy=dummy.rename(columns={'famsup_yes':'famsup'})
dummy=dummy.rename(columns={'paid_yes':'paid'})
dummy=dummy.rename(columns={'activities_yes':'activities'})
dummy=dummy.rename(columns={'internet_yes':'internet'})


dummy.head()
```

```python
dummy.set_index('Unnamed: 0', inplace=True)
```
